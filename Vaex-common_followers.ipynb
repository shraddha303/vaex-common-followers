## Count of Common followers in influencer

```python
import vaex
```

```python
import pandas as pd
import numpy as np
```

Creating data for influencer and followers
[i] In this data one column will be for followers id
[ii] 500 columns will be for influencer containing value as 0 or 1
[iii] 0 means follower is not following influencer on social media
[iv] 1 means follower is following influencer on social media

```python
n_rows = 1000000
n_cols = 500
```

```python
df = pd.DataFrame(np.random.choice([0, 1], size=(n_rows, n_cols)), columns=['inf%d' % i for i in range(n_cols)])
df.head()
```

```python
df['Followers_id'] =df.index
```

```python
df.info(memory_usage='deep')
```

## Writing data into csv

```python
file_path = 'Influencer_data1.csv'
df.to_csv(file_path, index=False)
```

## creating Hdf5 file

```python
vaex_df = vaex.from_csv(file_path, convert=True, chunk_size=5_000_000)
```
